```python
import pandas as pd
from sklearn.utils import shuffle
```

```python
recipes1 = pd.read_csv("RAW_recipes.csv", on_bad_lines='skip', engine='python')
```

```python
recipes1 = recipes1.drop(columns=['contributor_id', 'submitted', 'tags', 'nutrition', 'steps', 'description', 'id'])
```

```python
recipes1.info()
```

```python
recipes1.head()
```

```python
recipes1 = shuffle(recipes1)
```

```python
recipes2 = recipes1.iloc[:10000]
```

```python
recipes2 = recipes2.reset_index(drop=True)
```

```python
recipes2['id'] = recipes2.index
```

```python
recipes2
```

```python
recipes2 = recipes2[['id'] + [col for col in recipes2.columns if col != 'id']]
```
