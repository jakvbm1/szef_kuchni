```python
import pandas as pd
```

```python
df = pd.read_csv('recipes.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
all_ingredients = []
for ingredient_list in df['ingredients'].str.split(','):
    all_ingredients.extend(ingredient_list)
```

```python
import re
```

```python
clean = []
for ingredient in all_ingredients:
    clean.append(re.sub(r'[\[\]\'"]', '', ingredient).strip())
```

```python
len(clean)
```

```python
clean = list(set(clean))
```

```python
len(clean)
```

```python
df_of_ingredients = pd.DataFrame(clean, columns=['name'])
```

```python
df_of_ingredients.head(30)
```

```python


ingredients_linking_df = pd.DataFrame(df.ingredients.str.split(',').tolist(), index=df.name).stack().replace(r'[\[\]''"]', '', regex=True)
```

```python
ingredients_linking_df = ingredients_linking_df.reset_index([0, 'name'])
```

```python
ingredients_linking_df.columns = ['name', 'ingredient']
ingredients_linking_df = ingredients_linking_df.merge(df[['name', 'id']], on='name', how='left')
ingredients_linking_df.drop('name', axis=1, inplace=True)
ingredients_linking_df.rename(columns={'id': 'recipe_id'}, inplace=True)
ingredients_linking_df = ingredients_linking_df[["recipe_id", "ingredient"]]
ingredients_linking_df.rename(columns={'ingredient': 'name'}, inplace=True)
df_of_ingredients['index'] = df_of_ingredients.index
```

```python
ingredients_linking_df['name'] = ingredients_linking_df['name'].str.replace("'", "").str.strip().str.lower()
df_of_ingredients['name'] = df_of_ingredients['name'].str.strip().str.replace("'", "").str.lower()
```

```python
ingredients_linking_df = ingredients_linking_df.merge(
    df_of_ingredients[['name', 'index']],
    on='name',
    how='left'
)
```

```python

ingredients_linking_df.rename(columns={'index': 'ingredient_id'}, inplace=True)
```

```python
ingredients_linking_df['ingredient_id'].isna().sum()
```

```python
ingredients_linking_df.drop('name', axis=1, inplace=True)
```

```python
ingredients_linking_df["id"] = ingredients_linking_df.index

ingredients_linking_df = ingredients_linking_df[["id", "recipe_id", "ingredient_id"]]
ingredients_linking_df.to_csv('ingredients_linking.csv', index=False)
```

```python
df_of_ingredients = df_of_ingredients[["index", "name"]]
df_of_ingredients.to_csv('ingredients.csv', index=False)
```
